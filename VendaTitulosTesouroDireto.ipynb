```python
# Importando a biblioteca
import pandas as pd
```

```python
# Leitura do arquivo
df = pd.read_csv("/content/drive/MyDrive/colab/VendasTesouroDireto.csv", sep=";")
```

```python
# Exibindo as cinco primeiras linhas
df.head()
```

```python
# Renomeando a coluna PU
df = df.rename(columns={"PU":"Preco Unitario"})
```

```python
# Alterando o tipo de dados de determinadas colunas
df["Vencimento do Titulo"] = pd.to_datetime(df["Vencimento do Titulo"], format="%d/%m/%Y")
df["Data Venda"] = pd.to_datetime(df["Data Venda"], format="%d/%m/%Y")
df["Preco Unitario"] = df["Preco Unitario"].str.replace(",", ".").astype("float64")
df["Quantidade"] = df["Quantidade"].str.replace(",", ".").astype("float64")
df["Valor"] = df["Valor"].str.replace(",", ".").astype("float64")
```
